```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import seaborn as sns
```

```python
sns.countplot("FraudFound_P",data=df)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (18, 8))
df["Sex"].value_counts().plot.bar(color = "skyblue", ax = ax[0])
ax[0].set_title("Number Of People By Sex")
ax[0].set_ylabel("Population")
sns.countplot("Sex", hue = "FraudFound_P", data=df, ax = ax[1])
ax[1].set_title("Sex:Fraud vs Not-Fraud")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (18, 10))
df[df["FraudFound_P"] ==1]["Age"].plot.hist(ax = ax[0], bins = 20, edgecolor = "black", color = "blue")
ax[0].set_title("Not Fraud")
domain_1 = list(range(0, 85, 5))
ax[0].set_xticks(domain_1)
df[df["FraudFound_P"] == 0]["Age"].plot.hist(ax = ax[1], bins = 20, edgecolor = "black", color = "green")
ax[1].set_title("Fraud")
domain_2 = list(range(0, 85, 5))
ax[1].set_xticks(domain_2)
plt.show()
```
